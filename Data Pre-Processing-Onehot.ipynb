```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("./Data/Cleaned.csv")
```

```python
df.info()
```

```python
df['ClaimEvent_TPA_ID']=df['ClaimEvent_TPA_ID'].astype('object')
df['Reject status more than 3 months']=df['Reject status more than 3 months'].astype('object')
df['Class']=df['Class'].astype('object')
```

```python
df.info()
```

```python
df.drop('Medical Service Provider ID', axis=1,inplace=True)
```

```python
for features in df.select_dtypes(include = 'object').columns:
    print(features, ":", df[features].unique())
```

```python
X_values = df.iloc[:,:-1]
y_values = df.iloc[:,-1]
```

```python
categorical_features = X_values.select_dtypes(include = 'object').columns.values
```

```python
categorical_features
```

```python
X_onehot = pd.get_dummies(X_values,columns=categorical_features)
```

```python
# X_onehot.drop(categorical_features,axis=1, inplace=True)
print(X_onehot.columns.values)
```

```python
X_onehot
```

```python
#normalization of feature to bring the value in the range [0,1]
minMaxScale= MinMaxScaler() #minMax scaler
columns = X_onehot.columns
X_scaled = minMaxScale.fit_transform(X_onehot)
```

```python
X_scaled = pd.DataFrame(X_scaled, columns=columns)
X_scaled
```

```python
df_final = X_scaled
df_final['FraudFound'] = y_values
df_final.to_csv(r"./Data/df_final_preprocessed.csv", index=False)
```

```python
# Splitting the dataset into the Training set and Test set  
#to use sklearn we need data in matrix not in data frame
from sklearn.model_selection import train_test_split 
X = df_final.iloc[:, :-1]
y = df_final.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
X_train['FraudFound'] = y_train
X_test['FraudFound'] = y_test
```
